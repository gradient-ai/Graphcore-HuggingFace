# Training HuggingFace ðŸ¤— Models on an IPU using Paperspace Gradient

Whether you are looking to generate images with Stable Diffusion, derive insights from text, or need to recognize audio samples, the examples in here have you covered. And by using Paperspace to access IPUs you can be up and running in minutes!

## Getting Started with our interactive notebooks

In this runtime, you will be able to run models on IPUs for training and inference on vision transformers (ViT) for images, BERT-Large for question answering, and text classification, GPT-J for text generation and OpenAI's Whisper for speech to text conversion, all from within a Jupyter notebook.

You can find our notebooks in the folder directory on the left.
